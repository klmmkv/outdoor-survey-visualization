Research Question: 
Outdoor Knife Market Analysis: What is the purchasing and usage behaviour of an enthusiastic knife collector?

Due to the fact that I have great domain knowledge and consider myself a knife enthusiast, I posted my survey into a well-known Reddit group for fellow collectors. The population of the Reddit group is around one million users. In the end, I was able to collect 143 replies on my survey, which consisted of 8 questions. The survey was mostly based from the concept of "The Business Model Canvas" by Alexander Osterwalder. The questions aimed at separating people with knowledge about the topic, from people without, understanding which qualities people look for at purchase, and predict the willingness to pay. 

The survey consists of 7 questions and is attached in the appendix. The questions have been designed with a clear purpose, understand the purchasing and usage behaviour of an enthusiastic knife collector. The first two questions are based on a 6-point-likert scale and help us to identify motivated customers with knowledge, also called experts. 

1. I put great effort into selection of knives
2. I would consider myself to be a knife collector

Thus these questions help the survey to identify the more meaninigful or insightful answers. Next, we try to identify which qualities these people value when buying knifes (price, design, quality, ect.) 

3. Which knife would you buy? <br>
knife 1: (quality: very high, price: 799 dollars, production: exclusive, guarantee: lifetime)<br>
knife 2: (quality: good, price: 159 dollars, production: mass, guarantee: 5 years)<br>
4. What factors influenced your decision the most? (design, quality, price, production, guarantee) 

These two questions should give interesing insight into how to design the product and how to price it in order to convince the most people. Additionally we can check with the second question if the desire of people is consistent with their thoughts about their buying factors. 

The next question is for understanding what people buy the knives for and how much they are willing to pay for them:

5. Finish the sentence: I buy knifes for...  (collection, hunting, fun, safety, ... , etc. )
6. How much did you pay for your most expensive knife?

Last but not least, in order to cross-sell items, it would be smart to ask what additional items people buy when shopping for outdoor activities. 

7. Is there something else you buy related to outdoor activities?

```python
#extracting question about values and qualities extra from dataset

qualities = dataset.iloc[:,3]
qualities
```

```python
#visualizing question about values

f, ax = plt.subplots(figsize=(40, 20))
sns.countplot(qualities, order = qualities.value_counts().index)
plt.xticks(rotation=90);
```

This representation was chosen on purpose, as it might be the best visualization for this question to see the different combinations of values people consider when buying knives.
```python
#willingness to pay visualized

sns.histplot(dataset['How much did you pay for your most expensive knife?'])
```

This scale shows us a great indication of the willingness to pay. We can see that the main spectrum is above 100 dollars, with only around 11 people spending less then 100 dollars. It can be said that more then 50% spend more then 200$ on their most expensive knife - which shows that the willingness to pay is actually quite high. It might be possible to represent it in a form of a boxplot, to get a better representation of the distribution, however for this purpose the histogram is enough for now.

```python
#getting uses question extra

moreuses = dataset.iloc[:,4]
moreuses
```

```python
#creating a list of answers

moreuseslist = list(moreuses.values)
```

```python
moreuseslist
```

```python
#creating strings for better visualization further
string_of_words = "".join(str(moreuseslist))
```

```python
string_of_words
```

```python
#creating a wordcloud

from wordcloud import WordCloud

# Read the whole text.
text = string_of_words

# Generate a word cloud image
wordcloud = WordCloud().generate(text)

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt


# take relative word frequencies into account, lower max_font_size
wordcloud = WordCloud(background_color="white",max_words=len(string_of_words),max_font_size=40, relative_scaling=.5).generate(text)
plt.figure(figsize=(20,20))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

This was a quite difficult representation as it was an open ended question, nevertheless i believe a wordcloud is very suitable here in order to understand the main areas why people buy knives. From this wordcloud the main keywords which stand out are: Collection, Fun, Every Day Carry/Utility/Use (abbr. EDC). Moreover further areas might be: Safety, Hunting, Work, Camping. This was a great representation as it clearly shows that actually people buy knives for themselves (Collection and Fun) and for small everyday use, rather then for the real outdoors.
